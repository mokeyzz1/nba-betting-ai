```python
import joblib
import pandas as pd
from nba_api.stats.endpoints import LeagueDashTeamStats

# ‚úÖ Load trained model
model = joblib.load("xgboost_nba_model.pkl")
print("‚úÖ Model loaded successfully.")

# ‚úÖ Fetch latest NBA team stats
team_stats = LeagueDashTeamStats(season='2023-24', measure_type_detailed_defense='Base').get_data_frames()[0]

# ‚úÖ Ensure required columns exist
available_columns = team_stats.columns
print("‚úÖ Available columns in team_stats:", available_columns)

# ‚úÖ Select only relevant columns
selected_columns = ['TEAM_ID', 'TEAM_NAME', 'W', 'PACE', 'DEF_RATING', 'OFF_RATING', 'EFG_PCT']
team_stats = team_stats[selected_columns]

# ‚úÖ Load upcoming games (this should be a real data source)
upcoming_games = [
    {'home_team': 'Boston Celtics', 'away_team': 'Miami Heat'},
    {'home_team': 'Los Angeles Lakers', 'away_team': 'Golden State Warriors'}
]

# ‚úÖ Predict outcomes
for game in upcoming_games:
    home_team_stats = team_stats[team_stats['TEAM_NAME'] == game['home_team']].iloc[0]
    away_team_stats = team_stats[team_stats['TEAM_NAME'] == game['away_team']].iloc[0]

    # Prepare feature input
    game_data = pd.DataFrame([{
        'home_seasonWins': home_team_stats['W'],
        'away_seasonWins': away_team_stats['W'],
        'PACE': home_team_stats['PACE'],
        'DEF_RATING': home_team_stats['DEF_RATING'],
        'OFF_RATING': home_team_stats['OFF_RATING'],
        'EFG_PCT': home_team_stats['EFG_PCT']
    }])

    # Make prediction
    prediction = model.predict(game_data)
    winner = game['home_team'] if prediction[0] == 1 else game['away_team']

    print(f"üèÄ AI Prediction: {game['home_team']} vs. {game['away_team']} ‚Üí **Winner: {winner}** üéØ")

```
